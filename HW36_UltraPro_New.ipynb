Подходит для нормальной работы сразу только 3-НФ.

```python
# Создание таблицы customers
conn.execute('''
CREATE TABLE IF NOT EXISTS customers(
        customer_id numeric primary key,
        customer_name varchar(255),
        industry varchar(255)
        )''')
conn.commit()
```

```python
# Заполнение таблицы customers
conn.execute("INSERT INTO customers VALUES(1, 'Pupkin And Co', 'Agriculture')")
conn.execute("INSERT INTO customers VALUES(2, 'Tractor Plant', 'Auto-tractor Construction')")

conn.commit()
```

```python
# Создание таблицы customers_contact
conn.execute('''
CREATE TABLE IF NOT EXISTS customers_contact(
        customer_id numeric references customers(customer_id),
        customer_name varchar(255),
        contact_person_id integer,
        contact_person_and_role varchar(300),
        phone_number varchar(12),
        address varchar(255),
        city varchar(250),
        zip integer);''')
conn.commit()
```

```python
# Заполнение таблицы customers_contact
conn.execute("INSERT INTO customers_contact VALUES(1, 'Pupkin And Co', 123456, 'Paul McCartney, agronomist', '7915-670-123', 'Michigan str, 234', 'Michigan', 54321)")
conn.execute("INSERT INTO customers_contact VALUES(2, 'Tractor Plant', 789234, 'John Lennon, head engineer', '1216-345-321', '5-av, 234', 'NY', 13499)")
conn.commit()
```

Изменена таблица customer_project на project с исправлениями


```python
# Создание таблицы project, добавил project_id numeric primary key, связь с customer через customer_id numeric references customers(customer_id)
conn.execute('''
CREATE TABLE IF NOT EXISTS project(
        project_id numeric primary key, 
        customer_id numeric references customers(customer_id),
        customer_name varchar(255),
        project1_id integer,
        project1_feedback text,
        project2_id integer,
        project2_feedback text
        );''')
conn.commit()
```
