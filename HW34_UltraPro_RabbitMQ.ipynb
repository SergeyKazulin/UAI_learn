```python
import pika
import sys
import os
```

ОТПРАВКА СООБЩЕНИЯ. send.py

```python
# Установка соединения с сервером RabbitMQ
connect = pika.BlockingConnection(pika.ConnectionParameters('localhost'))   # подключение на локальном РС, иначе адрес или IP
channel = connect.channel()

# проверяем существование очереди получателей. Создадим очередь приветствия, в которую будет доставлено сообщение.
channel.queue_declare(queue='hello')
channel.basic_publish(exchange='',
                      routing_key='hello',
                      boby = 'Hello RabbitMQ!'
                      )

print('[x] Sent Hello RabbitMQ!')

# Выход из программы
connect.close()
```

Прием одной фнкцией receive.py

```python
def main():
    connect = pika.BlockingConnection(pika.ConnectionParameters('localhost'))   # подключение на локальном РС, иначе адрес или IP
    channel = connect.channel()

    channel.queue_declare(queue='hello')            # Установка соединения с сервером RabbitMQ

    def callback(ch, method, properties, body):     # Функция вывода содержимого сообщения на экран
        print('[x] Received %r' % body)
        print(f'[]x Принято: {body}')

    channel.basic_consume(queue='hello',
                      auto_ask=True,
                      on_message_callback=callback
                      )

    print('[*] Ожидание сообщений. \n Для выклбчения нажать CTRL+C')   # Режим ожидания
    channel.start_consuming()

if __name__ == '__main__':
    try:
        main()
    except KeyboardInterrupt:
        print('Соединение прервано')
        try:
            sys.exit(0)
        except SystemExit:
            os._exit(0)
```
