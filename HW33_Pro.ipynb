  «симфония», «симбиоз», «Симферополь», «Сим-карта», «синхрофазатрон» ,»Символизм»

```python
from pytrie import StringTrie 
  
def prefixSearch(arr, prefix): 
      
    trie = StringTrie() 
 
    for key in arr: 
        trie[key] = key 

    return trie.values(prefix) 
  
if __name__ == "__main__": 
    arr = ['симфония', 'симбиоз', 'Симферополь', 'Сим-карта', 'синхрофазотрон', 'Символизм']
    prefix = 'си'
    prefix_2 = 'Си'
    output = prefixSearch(arr, prefix)
    output_2 = prefixSearch(arr, prefix_2)
    if len(output) > 0: 
        print (sorted(output))
        print() 
        print(sorted(output_2))
    else: 
        print ('Шаблон не найден')
```
