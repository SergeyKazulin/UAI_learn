# 2 NF. 


```python
df_contacts_addresses = pd.DataFrame(columns=['contact_id',
                                'Full Names',
                               'Salutation',
                               'Physical Address',
                               ],
                       data=[
                             [1, 'Janet Jones', 'Ms.', 'First Street Plot N0 4'],
                             [2, 'Robert Phil', 'Mr.', '3rd Street 34'],
                             [3, 'Robert Phil', 'Mr.', '5th Avenue']
                             ])


df_rented = pd.DataFrame(columns=['rented_id',
                                  'Full Names',
                                  'Salutation',
                                  'Movies rented',
                               ],
                         data=[
                               [1, 'Janet Jones', 'Ms.', 'The girl next door'],
                               [2, 'Robert Phil', 'Mr.', 'Blade 3, Casablanca'],
                               [3, 'Robert Phil', 'Mr.', 'Titanic']
                               ])

print(df_contacts_addresses)
print()
print(df_rented)
```

# 3 NF.

```python
df_contacts = pd.DataFrame(columns=['contact_id',
                                'First Names',
                                'Last Name',
                                'Nickname',
                                'Salutation',
                               ],
                       data=[
                             [1, 'Janet', 'Jones', '', 'Ms.'],
                             [2, 'Robert', 'Phil', 'RPh', 'Mr.'],
                             [3, 'Robert', 'Phil', 'PhR', 'Mr.'],
                            ])

df_addresses = pd.DataFrame(columns=['address_id',
                                     'First Names',
                                     'Last Name',
                                     'Physical Address',
                               ],
                            data=[
                             [1, 'Janet', 'Jones', 'First Street Plot N0 4'],
                             [2, 'Robert', 'Phil', '3rd Street 34'],
                             [3, 'Robert', 'Phil', '5th Avenue']
                            ])

df_rented = pd.DataFrame(columns=['id_rented',
                                  'First Names',
                                  'Last Name',
                                  'Movie rented 1',
                                  'Movie rented 2'
                                 ],
                          data=[
                                [1, 'Janet', 'Jones','The girl next door'],
                                [2, 'Robert', 'Phil', 'Blade 3', 'Casablanca'],
                                [3, 'Robert', 'Phil', 'Titanic']
                                ])

print(df_contacts)
print()
print(df_addresses)
print()
print(df_rented)
```
